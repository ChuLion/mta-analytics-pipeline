```python
import pandas as pd
from google.cloud import bigquery

client = bigquery.Client(project="jdl-mta-project")

query = """
SELECT
    entries_delta,
    PERCENTILE_CONT(entries_delta, 0.25) OVER() as p25,
    PERCENTILE_CONT(entries_delta, 0.50) OVER() as p50,
    PERCENTILE_CONT(entries_delta, 0.75) OVER() as p75,
    PERCENTILE_CONT(entries_delta, 0.95) OVER() as p95,
    PERCENTILE_CONT(entries_delta, 0.99) OVER() as p99,
    PERCENTILE_CONT(entries_delta, 0.999) OVER() as p999,
    MAX(entries_delta) OVER() as max_delta
FROM `jdl-mta-project.mta_silver.stg_mta_turnstile_2019`
WHERE entries_delta IS NOT NULL
LIMIT 1
"""

df = client.query(query).to_dataframe()
print(df.T)
```

```python
query = """
SELECT
    APPROX_QUANTILES(median_household_income, 4) as quartiles,
    AVG(median_household_income) as mean_income,
    MIN(median_household_income) as min_income,
    MAX(median_household_income) as max_income
FROM `jdl-mta-project.mta_silver.stg_census_income`
WHERE median_household_income IS NOT NULL
"""
df = client.query(query).to_dataframe()
print(df)
```
